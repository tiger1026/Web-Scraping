```python
# we only need id, title, content, and excerpt 
# define a function to change parsed json to datafrome
def Scrape(ID):
    link = 'https://www.dcard.tw/_api/posts/' + str(ID)
    r = requests.get(link)
    js = r.json()
    return(pd.DataFrame(
        data=
        [{'ID':js['id'],
          'title':js['title'],
          'content':js['content'],
          'excerpt':js['excerpt'],
          'commentCount':js['commentCount']}], # keep comment count here since we might need it for the problem mentioned above
        columns=['ID','title','content','excerpt','commentCount']))
```

```python
# reading 100 posts from the popular section
url = 'https://www.dcard.tw/_api/posts?popular=true&limit=100'
r = requests.get(url)
js = r.json()
df = pd.DataFrame()
for i in range(len(js)):
    df = df.append(Scrape(js[i]['id']),ignore_index=True)
print(df.shape)
df
```
